1. What is a random variable in probability theory?
- A random variable in probability theory is a function that assigns a numerical value to each outcome in a sample space of a random experiment. It provides a way to quantify uncertain events in a structured mathematical manner.

2. What are the types of random variables?
- Random variables are classified into two main types based on the nature of their possible values:


*  Discrete Random Variable
*  Continuous Random Variable



3. What is the difference between discrete and continuous distributions?
- The difference between discrete and continuous distributions lies in the type of values the random variable can take and how probabilities are assigned.
-  Discrete Probability Distribution = Deals with discrete random variables (countable values).Probabilities are assigned to individual values using a probability mass function (PMF). Example: The number of heads in three coin tosses follows a Binomial Distribution with possible values X={0,1,2,3}.

- Continuous Probability Distribution = Deals with continuous random variables (uncountable values, typically real numbers in an interval).The total area under the PDF curve is 1. Heights of people often follow a Normal Distribution, where X can take any real number in a given range.

4. What are probability distribution functions (PDF)?
- A Probability Distribution Function (PDF) describes how probabilities are distributed over the possible values of a random variable. There are two main types of probability distribution functions:

- Probability Mass Function (PMF) - for Discrete Random Variables = The PMF gives the probability that a discrete random variable takes on a specific value. It is defined as: P(X=x)=f(x) where f(x) is the probability mass function.

- Probability Density Function (PDF) - for Continuous Random Variables = The PDF describes the likelihood of a continuous random variable falling within a given range. The probability of a single exact value is always zero: P(X=x)=0.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

- Probability Distribution Function (PDF/PMF) = Describes the likelihood of a random variable taking a specific value (PMF for discrete, PDF for continuous). PMF/PDF values can vary but must sum/integrate to 1. Tells how likely a value is to occur. Bell-shaped curve where f(x) represents density.

- Cumulative Distribution Function (CDF) = Describes the probability that a random variable is less than or equal to a given value. CDF gives cumulative probability, so F(x) is always increasing. CDF always ranges from 0 to 1. Tells the probability that a value is less than or equal to a given number. S-shaped curve where F(x) shows cumulative probability.

6. What is a discrete uniform distribution?
- A Discrete Uniform Distribution is a probability distribution where a finite number of outcomes are equally likely to occur. This means each possible value has the same probability. It's applications are: Games of chance (rolling dice, drawing a card from a deck), Randomized algorithms where equal probability selection is required, Simulations and Monte Carlo methods where uniform sampling is needed.

7. What are the key properties of a Bernoulli distribution?
- A Bernoulli distribution is a discrete probability distribution that represents a single trial of an experiment with only two possible outcomes: Success (X=1) with probability p, Failure (X=0) with probability 1−p.
It is the simplest form of a Binomial Distribution where only one trial is conducted.

8 . What is the binomial distribution, and how is it used in probability?
- The Binomial Distribution models the probability of obtaining a fixed number of successes in a fixed number of independent Bernoulli trials, where each trial has two possible outcomes: Success (1) with probability p, Failure (0) with probability 1−p.
It is used when an experiment is repeated multiple times under identical conditions, and we want to count the number of successful outcomes.


9. What is the Poisson distribution and where is it applied?
- The Poisson Distribution models the probability of a given number of events occurring in a fixed interval of time or space, assuming that:
- The events occur independently of each other.
- The average rate (λ) of occurrence is constant over time.
- Two events cannot happen at the exact same instant (rare event property).

10. What is a continuous uniform distribution?
- The Continuous Uniform Distribution is a probability distribution where all values in a given interval [a,b] are equally likely to occur. It is often called the rectangular distribution because its probability density function (PDF) is constant.

11.  What are the characteristics of a normal distribution?
- The Normal Distribution is one of the most important and widely used probability distributions in statistics. It is also known as the Gaussian distribution, and its bell-shaped curve has several key characteristics:
- Symmetry = The Normal Distribution is symmetric around its mean. This means the left and right sides of the distribution are mirror images of each other. The mean, median, and mode of a normal distribution are all equal and located at the center of the distribution.
- Bell-Shaped Curve = The shape of the normal distribution is a bell curve. It starts low, rises to a peak at the mean, and then falls symmetrically on both sides. It’s unimodal, meaning it has a single peak.
- Skewness and Kurtosis = A normal distribution has zero skewness, meaning it is perfectly symmetric. The kurtosis of a normal distribution is 3 (sometimes called mesokurtic). This describes the "peakedness" of the distribution; normal distributions have a moderate peak.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) is one of the most important and fundamental concepts in statistics. It states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables will approximate a normal distribution, regardless of the original distribution of the variables, provided the sample size is sufficiently large.
- Why is CLT Important in Statistics? = Foundation for Inference, Simplification of Complex Problems, Enables Use of the Normal Distribution, Large Sample Sizes, Application to Estimation.

14. How does the Central Limit Theorem relate to the normal distribution ?
- The Central Limit Theorem (CLT) and the normal distribution are closely connected because the CLT explains why the sampling distribution of the sample mean approximates a normal distribution as the sample size increases, regardless of the original distribution of the population.

15. What is the application of Z statistics in hypothesis testing ?
- Z-statistics (or z-scores) play a crucial role in hypothesis testing, particularly when the population variance is known, and the sample size is large. They help in determining whether there is enough statistical evidence to reject the null hypothesis based on sample data. In simple terms, Z-statistics allow us to compare the observed sample data to the expected data under the null hypothesis and assess if the difference is significant.

16.  How do you calculate a Z-score, and what does it represent ?
- Why is Z-Score Important?
- Standardization: Z-scores standardize data, making it easier to compare data points from different distributions. It allows you to compare how extreme or unusual a value is in different contexts (for example, comparing test scores from different exams).
- Comparing Different Datasets: Z-scores allow you to compare values across datasets with different means and standard deviations. For example, a score of 80 might be high in one exam, but low in another. The Z-score helps put it into context.
- Normal Distribution: In a normal distribution, about 68% of the data falls within 1 standard deviation from the mean (i.e., Z-scores between -1 and +1), 95% falls within 2 standard deviations (Z-scores between -2 and +2), and 99.7% falls within 3 standard deviations (Z-scores between -3 and +3). The Z-score tells you where the data point stands within this distribution.



17. What are point estimates and interval estimates in statistics ?
- In statistics, estimating unknown parameters (such as a population mean or proportion) is a fundamental task. There are two main ways to estimate these parameters: Point Estimate, Interval Estimate.
- Point Estimate = A point estimate is a single value or "point" that serves as the best guess for an unknown population parameter. It is based on sample data and represents the most likely value for that parameter. It is based on sample data, and may not exactly equal the population parameter due to random sampling variability. Point estimates are used when you need a single value to summarize or represent an unknown parameter.

- Interval Estimate = An interval estimate provides a range of values within which the unknown population parameter is likely to lie. This range is constructed around the point estimate, and it includes some measure of uncertainty or confidence. Interval estimates account for uncertainty and provide a range of plausible values, rather than a single value. A wider interval indicates more uncertainty about the parameter estimate, while a narrower interval indicates greater precision.

18. What is the significance of confidence intervals in statistical analysis ?
- Confidence intervals (CIs) play a central role in statistical analysis because they provide a range of values that estimates an unknown population parameter, such as a mean or proportion. Unlike point estimates, which offer a single value as an estimate, confidence intervals give a more informative range along with a confidence level that quantifies the uncertainty associated with the estimate.
- Provides a Range of Plausible Values
- Represents Uncertainty in Estimation
- Facilitates Decision-Making
- Indicates the Reliability of an Estimate
- Reduces the Impact of Outliers

19. What is the relationship between a Z-score and a confidence interval?

Z-score is used in hypothesis testing, standardizing data, and comparing scores from different normal distributions. Z-scores can be used to find probabilities in standard normal distributions. Z-scores are typically used in hypothesis testing and determining statistical significance.

- Confidence Interval = A Confidence Interval (CI) is a range of values, derived from sample data, that is used to estimate an unknown population parameter (e.g., mean or proportion) with a specific level of confidence. Confidence intervals are used to express the uncertainty or precision around an estimate of a population parameter. Confidence intervals give you a range of values and the confidence level (e.g., 95%) that the population parameter lies within this range. Confidence intervals are useful for parameter estimation, particularly when dealing with sample data.

20. How are Z-scores used to compare different distributions?
- Z-scores are used to compare different distributions by standardizing data points from different distributions. By converting raw scores into Z-scores, you can express them in terms of standard deviations from their respective means. This allows you to compare values from distributions with different means and standard deviations on the same scale, facilitating the assessment of how extreme or typical a particular value is across different datasets.

21. What are the assumptions for applying the Central Limit Theorem ?
- The assumptions for applying the Central Limit Theorem (CLT) are:

1.   Independence: The observations in the sample must be independent of each other. This means the selection of one data point should not affect the selection of another.
2.   Sample Size: The sample size must be sufficiently large. Although there is no strict rule, a common guideline is that the sample size n should be at least 30 for the CLT to hold. For highly skewed distributions, larger sample sizes might be needed.
3. Random Sampling: The data should be obtained through random sampling to ensure that every individual in the population has an equal chance of being selected, and the sample represents the population well.
4. Underlying Distribution: The CLT applies to both normal and non-normal distributions. For non-normal distributions, the sample size should be large enough to ensure the sampling distribution of the sample mean becomes approximately normal.



22. What is the concept of expected value in a probability distribution ?

It represents the center of the distribution and is calculated as the sum of each possible outcome multiplied by its probability. The expected value is denoted by
E(X) and is a measure of the "center" of the distribution.

The expected outcome of a random variable is the weighted average of all possible outcomes, where each outcome is weighted by its corresponding probability, as defined by the probability distribution.
```python
# Q9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)

    z_scores = (data - mean) / std_dev

    return z_scores
data = np.random.normal(0, 1, 1000)

z_scores = calculate_z_scores(data)

plt.plot(z_scores, marker='o', linestyle='-', color='skyblue', label='Z-scores')
plt.axhline(0, color='black', linewidth=1)
plt.xlabel('Index')
plt.ylabel('Z-score')
plt.title('Z-scores of the Dataset')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Q15.  Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom

def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)

    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.title('Standard Normal Distribution (Mean = 0, Std = 1)')
    plt.legend()
    plt.grid()
    plt.show()

def binomial_distribution(n, p, size):
    x = np.arange(0, n+1)
    probabilities = binom.pmf(x, n, p)

    random_vars = binom.rvs(n, p, size=size)

    plt.figure(figsize=(8, 5))
    plt.bar(x, probabilities, color='blue', alpha=0.6, label='Binomial PMF')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.legend()
    plt.grid()
    plt.show()

    return random_vars

def calculate_z_score(data_point, mean, std_dev):
    return (data_point - mean) / std_dev

def hypothesis_testing(sample_mean, population_mean, std_dev, sample_size, alpha=0.05):
    z_score = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    print(f'Z-score: {z_score}')
    print(f'P-value: {p_value}')

    if p_value < alpha:
        print("Reject the null hypothesis: There is significant evidence to support the alternative hypothesis.")
    else:
        print("Fail to reject the null hypothesis: There is not enough evidence to support the alternative hypothesis.")

plot_standard_normal()
binomial_distribution(n=10, p=0.5, size=1000)

data_point = 1.5
mean = 0
std_dev = 1
z_score = calculate_z_score(data_point, mean, std_dev)
print(f'Z-score for data point {data_point}: {z_score}')

hypothesis_testing(sample_mean=5.2, population_mean=5, std_dev=1, sample_size=30)
```

```python
# Q16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, t

def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)

    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.title('Standard Normal Distribution (Mean = 0, Std = 1)')
    plt.legend()
    plt.grid()
    plt.show()

def binomial_distribution(n, p, size):
    x = np.arange(0, n+1)
    probabilities = binom.pmf(x, n, p)

    random_vars = binom.rvs(n, p, size=size)
    plt.figure(figsize=(8, 5))
    plt.bar(x, probabilities, color='blue', alpha=0.6, label='Binomial PMF')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.legend()
    plt.grid()
    plt.show()

    return random_vars

def calculate_z_score(data_point, mean, std_dev):
    return (data_point - mean) / std_dev

def hypothesis_testing(sample_mean, population_mean, std_dev, sample_size, alpha=0.05):
    z_score = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    print(f'Z-score: {z_score}')
    print(f'P-value: {p_value}')

    if p_value < alpha:
        print("Reject the null hypothesis: There is significant evidence to support the alternative hypothesis.")
    else:
        print("Fail to reject the null hypothesis: There is not enough evidence to support the alternative hypothesis.")

def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)
    margin_of_error = t.ppf((1 + confidence) / 2, df=n-1) * (sample_std / np.sqrt(n))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f'{confidence*100}% Confidence Interval: ({lower_bound}, {upper_bound})')
    return lower_bound, upper_bound

plot_standard_normal()
binomial_distribution(n=10, p=0.5, size=1000)

data_point = 1.5
mean = 0
std_dev = 1
z_score = calculate_z_score(data_point, mean, std_dev)
print(f'Z-score for data point {data_point}: {z_score}')

hypothesis_testing(sample_mean=5.2, population_mean=5, std_dev=1, sample_size=30)

data_sample = np.random.normal(50, 10, 30)
confidence_interval(data_sample)
```

```python
# Q21. Write a Python program to compare the standard deviations of two datasets and visualize the difference.


import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=1000)
data2 = np.random.normal(loc=50, scale=20, size=1000)

std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(data1, bins=30, alpha=0.5, label=f"Dataset 1 (Std={std1:.2f})", color='blue')
axes[0].hist(data2, bins=30, alpha=0.5, label=f"Dataset 2 (Std={std2:.2f})", color='red')
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")
axes[0].set_title("Histogram of Two Datasets")
axes[0].legend()
axes[0].grid(True)

axes[1].boxplot([data1, data2], vert=True, patch_artist=True,
                labels=[f"Dataset 1 (Std={std1:.2f})", f"Dataset 2 (Std={std2:.2f})"],
                boxprops=dict(facecolor="lightblue"),
                medianprops=dict(color="black"))
axes[1].set_title("Boxplot Comparison")

plt.tight_layout()
plt.show()
```

```python
# Q23. Implement Z-score normalization on a dataset and visualize its transformation.


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)

normalized_data = zscore(data)

original_mean = np.mean(data)
original_std = np.std(data, ddof=1)
normalized_mean = np.mean(normalized_data)
normalized_std = np.std(normalized_data, ddof=1)

print(f"Original Mean: {original_mean:.2f}, Original Std Dev: {original_std:.2f}")
print(f"Normalized Mean: {normalized_mean:.2f}, Normalized Std Dev: {normalized_std:.2f}")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='blue', alpha=0.7, label="Original Data")
plt.axvline(original_mean, color='red', linestyle='dashed', linewidth=2, label="Mean")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Original Data")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='green', alpha=0.7, label="Normalized Data")
plt.axvline(normalized_mean, color='red', linestyle='dashed', linewidth=2, label="Mean")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-score Normalized Data")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#  Q24.  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.


import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):

    skewness = skew(data)
    kurt = kurtosis(data)

    return skewness, kurt

np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)

skewness, kurt = calculate_skewness_kurtosis(data)

print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis (Excess): {kurt:.4f}")
```
